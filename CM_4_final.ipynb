## HEART DISEASE DATASET

#### IMPORTING LIBRAIRES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

#### IMPORT HEART DISEASE DATASET

```python
dataset = pd.read_csv("heart_disease_missing.csv")
dataset['thal'] = dataset['thal'].round()
dataset.dropna(subset = ['sex','age','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal'], inplace=True)
dataset.describe()
```

#### HISTOGRAM TO SHOW PRESENT AND ABSENT HEART DISEASE CASES

```python
fig = plt.figure(figsize=(15,10))

plt.subplot(231)
plt.title("Distribution of chest pain")
sns.countplot(x='cp', hue='target', data = dataset, palette = 'dark')

plt.subplot(232)
plt.title("Distribution of first blood sugar")
sns.countplot(x='fbs', hue='target', data = dataset, palette="Set1")

plt.subplot(233)
plt.title("Distribution of sex")
sns.countplot(x='sex', hue='target', data = dataset, palette = 'dark')

plt.subplot(234)
plt.title("Distribution of Resting ECG")
sns.countplot(x='restecg', hue='target', data = dataset, palette = 'Set1')

plt.subplot(235)
plt.title("Distribution of exercise induced angina")
sns.countplot(x='exang', hue='target', data = dataset, palette = 'dark')

plt.subplot(236)
plt.title("Distribution of slope")
sns.countplot(x='slope', hue='target', data = dataset, palette = 'Set1')
```

<h2><center>key features observed from histogram plot over target</center></h2>
<ul>
            <li>-->It can be observed that chest pain of type(1 ,2 and 3) is more common with people having heart disease.      Therefore, chest pain is a dominant feature.</li>
            <li>-->It can be observed that people with heart disease have almost equal distribution for high and low fasting blood sugar.Therefore,this feature has less correlation with target</li>
            <li>-->Heart disease is more prominant in females in comparision to males. </li>
            <li>-->People with no exercise induced angina have more heart related disease compared to people with exercise induced angina.</li>
            <li>-->Flat slope line clearly shows a higher ratio of people with heart related disease.</li>
            <li>-->It can be seen that resting ECG has a very weak correlation with target as no particular point is dominant for heart related disease.</li>
</ul>

```python
fig = plt.figure(figsize=(15,10))

plt.subplot(211)
sns.countplot(x='age', hue='target', data = dataset, palette = 'dark')

plt.subplot(212)
sns.countplot(x='thal', hue='target', data = dataset, palette = 'dark')
```

<h2><center>key features observed from histogram plot over target</center></h2>
<ul>
            <li>-->There is no clear observation from age histogram. However, It can be observed that the majority of the people have age less than 55 have shown the onset of heart disease compared to people with age more than 55 years old.</li>
            <li>-->People with normal Thalium stress test results have suffered a heart disease compared to people with reversible defect</li>
</ul>
