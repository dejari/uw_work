## IRIS DATASET

#### IMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()

from scipy.stats import skew
from scipy.stats import kurtosis
import statistics
```

#### IMPORTING DATASET AND DESCRIPTION OF THE DATA

```python
dataset = pd.read_csv("iris_dataset_missing.csv")
dataset = dataset.dropna(axis = 0)
dataset.describe()
```

#### CORRELATION, MEAN, VARIANCE , SKEW AND KURTOSIS FOR THE DATASET

```python
# checking the correlation of data
correlation_data = dataset.corr()
sns.heatmap(correlation_data, cmap='magma', annot= True, cbar= True)
print(correlation_data)
```

<h4><center>The above box shows a correlation table and a heat map among different features in the dataset.</center></h4>
<ul>
            <li>-->All the features are strongly correlated with each other except sepal-width.</li>
            <li>-->The highest correlation is observed between petal-length and petal-width, which indicates a strong linear                       relation among the variables.</li>
            <li>-->A negative correlation is observed between sepal-width and other features, which indicates an inverse relation                 among variables.</li>
</ul>

```python
fig = plt.figure(figsize=(15,10))

plt.subplot(221)
plt.title("Distribution of sepal-width")
sns.distplot(dataset['sepal_width'], hist = True, kde = True)


plt.subplot(222)
plt.title("Distribution of sepal-length")
sns.distplot(dataset['sepal_length'], hist = True, kde = True)

plt.subplot(223)
plt.title("Distribution of petal-width")
sns.distplot(dataset['petal_width'], hist = True, kde = True)

plt.subplot(224)
plt.title("Distribution of petal-length")
sns.distplot(dataset['petal_length'], hist = True, kde = True)

print(f"mean of the sepal_width data: {statistics.mean(dataset['sepal_width'])}")
print(f"variance of the sepal_width data: {statistics.variance(dataset['sepal_width'])}")
print(f"Skewness of the sepal_width data: {skew(dataset['sepal_width'])}")
print(f"kurtosis of the sepal_width data: {kurtosis(dataset['sepal_width'])}")
print("\n")
print(f"mean of the sepal_length data: {statistics.mean(dataset['sepal_length'])}")
print(f"variance of the sepal_length data: {statistics.variance(dataset['sepal_length'])}")
print(f"Skewness of the sepal_length data: {skew(dataset['sepal_length'])}")
print(f"kurtosis of the sepal_length data: {kurtosis(dataset['sepal_length'])}")
print("\n")      
print(f"mean of the petal_length data: {statistics.mean(dataset['petal_length'])}")
print(f"variance of the petal_length data: {statistics.variance(dataset['petal_length'])}")
print(f"Skewness of the petal_length data: {skew(dataset['petal_length'])}")
print(f"kurtosis of the petal_length data: {kurtosis(dataset['petal_length'])}")
print("\n")      
print(f"mean of the petal_width data: {statistics.mean(dataset['petal_width'])}")
print(f"variance of the petal_width data: {statistics.variance(dataset['petal_width'])}")
print(f"Skewness of the petal_width data: {skew(dataset['petal_width'])}")
print(f"kurtosis of the petal_width data: {kurtosis(dataset['petal_width'])}")
```

<h4><center>Key features observed in the distribution and statistics of the data.</center></h4>
<ul>
            <li>-->The sepal-width distribution shows a positive skew with less magnitude that indicates data is still symmetrical.  This distribution is leptokurtic, and it observes a flaw around the range of (3.5,4) that indicates a presence of outliers. Overall the data is almost distributed symmetrical along the mean. Majority of the distribution of sepal-width is between 2.5cm to 3.5cm.</li>
            <li>-->The sepal-length distribution shows a positive skew with significantly less magnitude that indicates data is very close to normal distribution. This distribution is platykurtic, and it observes a break in symmetry around the range of (6,7) and lack of smoothness of the curve.</li>
            <li>-->The petal-length distribution shows a negative skew with less magnitude and has two peaks. This distribution is platykurtic, and it covers the highest range of values and almost bimodal distribution.</li>
             <li>-->The petal-width distribution shows a negative skew with a magnitude of almost zero that indicates data is very close to normal distribution. This distribution is platykurtic, and it observes a flaw around the range of (0.5,1) giving the distribution two peaks. Overall the data for petal-width is closest to the normal distribution compared to other features.</li>
</ul>
